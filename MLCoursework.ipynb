```
#Problem One - Linear regression 

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("airbnb_dataset.csv")
X= df['Type_of_room'] = df.room_type.map({'Private room': 0, 'Entire home/apt':1, 'Shared room':2}) 
#^^ creating a new column in dataset using data from the 'room type' column and assigned numbers to the text. 
y= df['price']

X= np.array(df['Type_of_room']).reshape(-1, 1) #changing the shape of an array.
y = np.array(df['price'])

xtrain, xtest, ytrain, ytest = train_test_split(X,y,test_size=1/3, random_state=0)
#Splitting data into training a testing data. 

regressor = LinearRegression() #creates the variable model as the instance of LinearRegression
regressor.fit(xtrain, ytrain) #calculate the optimal values of x and y 
pred = regressor.predict(xtest) # predict the price of rooms 

r_sq = regressor.score(X, y)
print('coefficient of determination:', r_sq) #score to evaluate the performance of the model. 
print('intercept:', regressor.intercept_) #expected mean on y when x=0 
print('slope:', regressor.coef_) #slope, how much you except y to change when x increases 
y_pred = regressor.predict(X) 
print('predicted response:', y_pred, sep='\n') #prints the predictions of prices of any room type  

#Scatter plot of the training results 
plt.scatter(xtrain, ytrain, color= 'orange')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title ("Training Dataset")
plt.xlabel("Room type")
plt.ylabel("Price")
plt.show()

#Scatter plot of the Test Results 
plt.scatter(xtest, ytest, color= 'green')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title("Test DataSet")
plt.xlabel("Room Type")
plt.ylabel("Price")
plt.show()
 
# average of the square of the difference between the observed and predicted values of a variable
print('Mean squared error: %.2f'
      % mean_squared_error(ytest,pred))
```

```
import pandas as pd
import numpy as np
import random as rd
import matplotlib.pyplot as plt

data = pd.read_csv('airbnb_dataset.csv')
data.head()
```

```
#Visualising the number of reviews against the price 
X = data[["number_of_reviews","price"]]
plt.scatter(X["number_of_reviews"],X["price"],c='blue')
plt.xlabel('number of reviews')
plt.ylabel('price of room')
plt.show()
```

```
K=3 #The number of clusters, this was calucated using the elbow method. 

# Select random observation as centroids - they are picked at random everytime you run the code the point change 
Centroids = (X.sample(n=K))
plt.scatter(X["number_of_reviews"],X["price"],c='black')
plt.scatter(Centroids["number_of_reviews"],Centroids["price"],c='red')
plt.xlabel('Number of reviews')
plt.ylabel('Price of room')
plt.show()
```

```
diff = 1
j=0

while(diff!=0):
    XD=X
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["number_of_reviews"]-row_d["number_of_reviews"])**2
            d2=(row_c["price"]-row_d["price"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        X[i]=ED
        i=i+1

    C=[]
    for index,row in X.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(K):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+1
        C.append(pos)
    X["Cluster"]=C
    Centroids_new = X.groupby(["Cluster"]).mean()[["price","number_of_reviews"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['price'] - Centroids['price']).sum() + (Centroids_new['number_of_reviews'] - Centroids['number_of_reviews']).sum()
        print(diff.sum())
    Centroids = X.groupby(["Cluster"]).mean()[["price","number_of_reviews"]]
```

```
color=['blue','green','cyan']
for k in range(K):
    data=X[X["Cluster"]==k+1]
    plt.scatter(data["number_of_reviews"],data["price"],c=color[k])
plt.scatter(Centroids["number_of_reviews"],Centroids["price"],c='red')
plt.xlabel('number of reviews')
plt.ylabel('price of room')
plt.show()
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.cluster import KMeans

df = pd.read_csv("airbnb_dataset.csv")

df.head()

x = df.iloc[:, [9,11]].values #selecting the 9th (price) and 11th (number of reviews) column 
```

```
from sklearn.cluster import KMeans
wcss = [] #Within-Cluster-Sum-of-Squares

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
#Data visuilsation 
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```
